## 2) Description from kaggle about the different data types:

The data consists of 100,000 observations of space taken by the SDSS (Sloan Digital Sky Survey). Every observation is described by 17 feature columns and 1 class column which identifies it to be either a star, galaxy or quasar.

obj_ID = Object Identifier, the unique value that identifies the object in the image catalog used by the CAS

alpha = Right Ascension angle (at J2000 epoch)

delta = Declination angle (at J2000 epoch)

u = Ultraviolet filter in the photometric system

g = Green filter in the photometric system

r = Red filter in the photometric system

i = Near Infrared filter in the photometric system

z = Infrared filter in the photometric system

run_ID = Run Number used to identify the specific scan

rereun_ID = Rerun Number to specify how the image was processed

cam_col = Camera column to identify the scanline within the run

field_ID = Field number to identify each field

spec_obj_ID = Unique ID used for optical spectroscopic objects (this means that 2 different observations with the same spec_obj_ID must share the output class)

class = object class (galaxy, star or quasar object)

redshift = redshift value based on the increase in wavelength

plate = plate ID, identifies each plate in SDSS

MJD = Modified Julian Date, used to indicate when a given piece of SDSS data was taken

fiber_ID = fiber ID that identifies the fiber that pointed the light at the focal plane in each observation

- From this description we can identify some of the features which are not necessary for our classification (**requires physical intuition**).

- This works under the assumption that the are no flaws involved in the process of actually gathering the data.

- Even if we dont fully understand each feature (would require more detailed knowledge about the physical implications at hand) we should later be able to identify further features which arent well suited for usage in the classification (by looking at hit and miss rate for each feature for example).

Since our data is quite imbalanced we want to resample it to 30000 representatives for each class.

```python
CLASSES = ['GALAXY', 'QSO', 'STAR']

df_sampled=pd.DataFrame()

for class_ in CLASSES:
    df_sample = df[df[class_] == 1]
    
    if len(df_sample) >= 30000:
        df_sample = df_sample.sample(n = 30000, axis = 0)
    else:
        df_add = df_sample.sample(n = 30000 - len(df_sample), axis = 0)
        df_sample = pd.concat([df_add, df_sample], axis = 0)
    
    df_sampled = pd.concat([df_sampled, df_sample], axis = 0, ignore_index = True)
    
print(df_sampled.shape)
for class_ in CLASSES:
    print(len(df_sampled[df_sampled[class_] == 1]))
```

With this sampling method, we oversample our underrepresented classes and undersample the overrepresented ones.

This short script can easily be generalized to work with an arbritrary, statistically similiar, dataset.

```python
from torch.utils.data import Dataset, TensorDataset

class CustomDataset(Dataset):
    def __init__(self, x_tensor: torch.tensor, y_tensor: torch.tensor) -> None:
        self.x = x_tensor
        self.y = y_tensor
        
    def __getitem__(self, index):
        return (self.x[index], self.y[index])

    def __len__(self):
        return len(self.x)
```

# 8) Build and train model

```python
from torch import nn
```

```python
INPUT_DIM = 7
OUTPUT_DIM = 3

import torch.nn.functional as F

class StellarClassification(nn.Module):
    def __init__(self, input_features = INPUT_DIM, output_features = OUTPUT_DIM, hidden_units = 16):
        
        assert(hidden_units > 1), "Hidden units <= 1!"
        
        super(StellarClassification, self).__init__()
        
        self.linear1 = nn.Linear(in_features = input_features, out_features = hidden_units)

        self.linear2 = nn.Linear(in_features = hidden_units, out_features = hidden_units)

        self.linear3 = nn.Linear(in_features = hidden_units, out_features = output_features)

        self.linear4 = nn.Linear(in_features = output_features, out_features = output_features)
        
        
    def forward(self, x):
        out = self.linear1(x)
        out = F.tanh(out)
        out = self.linear2(out)
        out = F.relu(out)
        out = self.linear3(out)
        out = F.relu(out)
        out = self.linear4(out)
        return out
```

```python
model = StellarClassification(INPUT_DIM, OUTPUT_DIM, hidden_units = 10).to(device)
```

```python
print(model)
print(next(model.parameters()).is_cuda)
```

```python
loss_fn = nn.CrossEntropyLoss()
```

```python
optimizer = torch.optim.Adam(params = model.parameters())
```

Setup for training routine:

```python
def make_train_step(model, loss_fn, optimizer):
    # Builds function that performs a step in the train loop
    def train_step(x, y):
        # Sets model to TRAIN mode
        model.train()
        # Makes predictions
        yhat = model(x)
        # Computes loss
        loss = loss_fn(yhat, y)
        # Computes gradients
        optimizer.zero_grad()
        loss.backward()
        # Updates parameters and zeroes gradients
        optimizer.step()
        # Returns the loss
        return (yhat, loss.item())
    
    # Returns the function that will be called inside the train loop
    return train_step
```

Adding early stopping routine:

```python
class train_tracker():
    def __init__(self, model_path = '', patience = 4, difference = 0.05):
        self.metric = 0 # if higher is better set low, otherwise set high
        self.model_path = model_path
        self.patience = patience
        self.difference = difference
        self.count = self.patience
        self.best = None
    
    def save_model(self, best_params, model_path):
        try:
            torch.save(best_params, model_path)
        except:
            print('Track Error!')
    
    def call(self, metric):
        if ((self.metric - self.difference) < metric) or (self.count > 0):
            if (self.metric - self.difference) < metric:
                self.metric = metric
                self.count = self.patience
                try:
                    self.best = model.state_dict()
                except:
                    print('Best Error!')
                return True
            else:
                self.count = self.count - 1
                return True
        else:
            self.save_model(best_params = self.best, model_path = self.model_path)
            return False
```

```python
import torchmetrics
import os

def train_model(model, epochs, loss_fn, optimizer, model_name, patience, difference):
    epochs = 50

    losses = []
    val_losses = []
    train_step = make_train_step(model, loss_fn, optimizer)
    accs = [] # per batch
    val_accs = [] #per batch

    tracker = train_tracker(model_path = os.path.join('models', model_name), patience = patience, difference = difference)

    for epoch in range(epochs):
        n_samples = 0
        n_correct = 0
        for x_batch, y_batch in train_loader:
            x_batch = x_batch.to(device)
            y_batch = y_batch.to(device)

            y_hat, loss = train_step(x_batch, y_batch)
            losses.append(loss)

            _, predictions_1 = torch.max(y_batch, dim = 1)
            _, predictions_2 = torch.max(y_hat, dim = 1)

            n_samples += y_batch.shape[0]
            n_correct += (predictions_1 == predictions_2).sum().item()

            acc = n_correct/n_samples
            accs.append(acc)

        n_samples = 0
        n_correct = 0
        with torch.no_grad():
            for x_val, y_val in val_loader:
                x_val = x_val.to(device)
                y_val = y_val.to(device)

                model.eval()

                y_hat = model(x_val)
                val_loss = loss_fn(y_hat, y_val)

                val_losses.append(val_loss.item())

                _, predictions_1 = torch.max(y_val, dim = 1)
                _, predictions_2 = torch.max(y_hat, dim = 1)

                n_samples += y_val.shape[0]
                n_correct += (predictions_1 == predictions_2).sum().item()

                val_acc = n_correct/n_samples
                val_accs.append(val_acc)

            if not tracker.call(metric = val_accs[-1]):
                break

        if (epoch % 1 == 0) or (epoch == epochs-1):
            print(f"Epoch: {epoch} | Loss: {losses[-1]:.5f}, Accuracy: {accs[-1]*100:.2f}% | Val loss: {val_losses[-1]:.5f}, Validation accuracy: {val_accs[-1]*100:.2f}%")
    
    return (*[losses, accs, val_losses, val_accs],)

losses, accs, val_losses, val_accs = train_model(model=model, epochs=120, loss_fn=loss_fn, optimizer=optimizer,
                                                 model_name='trained_model', patience=5, difference=0.)
```

```python
metrics = [losses, accs, val_losses, val_accs]
metric_names = ['Training loss', 'Training accuracy', 'Validation loss', 'Validation accuracy']
```

```python
fig, axs = plt.subplots(nrows = len(metrics), figsize = (12, 10))
for i, metric in enumerate(metrics):
    axs[i].plot([i+1 for i in range(len(metric))], metric)
    axs[i].set_title(metric_names[i])
    axs[i].set_xlabel('Batches')
    axs[i].set_ylabel(metric_names[i])

fig.tight_layout()
plt.show()
```

# 9) First model evaluation

```python
def convert_to_binary(y_hat: np.ndarray) -> np.ndarray:
    for i in range(len(y_hat)):
        max_element = np.argmax(y_hat[i])
        for j in range(len(y_hat[i])):
            if j == max_element:
                y_hat[i,j]=1.
            else:
                y_hat[i,j]=0.
    return y_hat
```

```python
import sklearn.metrics

def plot_cm(model, test_loader):
    model.eval()
    y_hat_test = None
    y_test = None
    x_test = None
    
    for x_test, y_test in test_loader:
        x_test = x_test.to(device)
        y_hat_test = model(x_test)

    y_hat_test = y_hat_test.cpu().detach().numpy()
    y_hat_test = convert_to_binary(y_hat_test)

    try: y_test = y_test.cpu().detach().numpy()
    except: print('y_test already on cpu')

    cm_GALAXY=sklearn.metrics.confusion_matrix(y_true=y_test[:,0], y_pred=y_hat_test[:,0])
    dp_GALAXY=sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=cm_GALAXY)
    cm_STAR=sklearn.metrics.confusion_matrix(y_true=y_test[:,1], y_pred=y_hat_test[:,1])
    dp_STAR=sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=cm_STAR)
    cm_QSO=sklearn.metrics.confusion_matrix(y_true=y_test[:,2], y_pred=y_hat_test[:,2])
    dp_QSO=sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=cm_QSO)

    dp_GALAXY.plot()
    dp_STAR.plot()
    dp_QSO.plot()
    
    acc_G=np.trace(cm_GALAXY)/np.sum(cm_GALAXY)
    print(f"Accuracy of identifying Galaxies: {acc_G*100:.3f}%")
    acc_S=np.trace(cm_STAR)/np.sum(cm_STAR)
    print(f"Accuracy of identifying Stars: {acc_S*100:.3f}%")
    acc_Q=np.trace(cm_QSO)/np.sum(cm_QSO)
    print(f"Accuracy of identifying Quasars: {acc_Q*100:.3f}%")
    
    mean_acc = (acc_G + acc_S + acc_Q)/3.
    print(f'Mean accuracy across classes: {mean_acc*100:.3f}%')
    
    try: x_test = x_test.cpu().detach().numpy()
    except: print('X_test already on cpu')
    
    return (*[x_test, y_test, y_hat_test],)
```

```python
X_test, y_test, y_hat_test = plot_cm(model, test_loader)
```

Looking at performance per feature:

```python
def feature_plot(X_test, y_test, y_hat_test):
    X_hit=[]
    X_miss=[]

    try: X_test = X_test.cpu().detach().numpy()
    except: print('X_test was already on cpu')

    for i in range(len(y_hat_test)):
        if np.array_equal(y_hat_test[i], y_test[i]):
            X_hit.append(X_test[i,:])
        else:
            X_miss.append(X_test[i,:])
    
    # changed dtype because of possible 0 standard deviations
    X_hit=np.array(X_hit, dtype=np.float64)
    X_miss=np.array(X_miss, dtype=np.float64)
    
    print(f'Hits: {len(X_hit)/len(X_test)*100:.4f}%')
    # print(f'Hit stds: {X_hit.std(axis=0)}')
    x_k_hit=X_hit.mean(axis=0)/X_hit.std(axis=0)
    
    print(f'Misses: {len(X_miss)/len(X_test)*100:.4f}%')
    # print(f'Miss stds: {X_miss.std(axis=0)}')
    # print(f'{X_miss}')
    x_k_miss=X_miss.mean(axis=0)/X_miss.std(axis=0)

    plt.figure(figsize=(6,4), dpi=150)
    plt.plot([i for i in range(len(x_k_hit))], x_k_hit, ".--r", label=r"$\overline{x}_{k,hit}/\sigma_{k,hit}$")
    plt.plot([i for i in range(len(x_k_hit))], x_k_miss, "x--b", label=r"$\overline{x}_{k,miss}/\sigma_{k,miss}$")
    plt.xlabel("k-th feature")
    plt.ylabel(r"$\overline{x}_k/\sigma_k$")
    plt.grid()
    plt.legend()
    plt.show()
```

```python
feature_plot(X_test, y_test, y_hat_test)
```

# 10) Further improvement (automated outlier detection)
After this first model prototype, I experimented with different hyperparameters and bachsizes. The accuracies never really improved, this could mean that there are still outliers in our data.

```python
classes = ['GALAXY', 'QSO', 'STAR']
df.head()
```

```python
df.drop(classes + ['spec_obj_ID'], axis = 1).plot(kind = 'box')
plt.show()
```

```python
df['spec_obj_ID'].plot(kind = 'box')
plt.show()
```

As we can see, at least according the box plot criterium (values lying in the intervall [Q1 - 1.5 IQR, Q3 + 1.5 IQR]), there are definetly some outliers left in our dataset.

Box plots are a realiable way of identifying outliers in a single dimension feature space. If we also want to take into account the dependencies of the feature variables we need other methods for identifying outliers. One such way is through the use of Isolation forests, they allow predicting outliers through implying a sence of distance (or isolation) of feature points in the high dimensional feature space.

Since our training did not allow for higher accuracys than 96% we will assume an outlier contamination of 4%.

```python
from sklearn.ensemble import IsolationForest

model_IF = IsolationForest(contamination = 0.04, random_state = 42)

df_ = df_sampled.drop(classes, axis = 1)

model_IF.fit(df_)
```

Add outlier score (negative means outlier, lower means higher certainty of outlier):

```python
df_['outlier_scores'] = model_IF.decision_function(df_sampled.drop(classes, axis = 1))
```

Add outlier prediction (-1 means outlier 1 means inlier):

```python
df_['outlier'] = model_IF.predict(df_sampled.drop(classes, axis = 1))
```

# 11) Summary
**By introducing a feature which was motivated based on pyhsics and removing outliers, I increased the performance of the model from roughly 97.7% to almost 100%.**

This short project showed how phyiscal intuition help with the following problems:
 - Feature selection: Identifying features which carry relevant and not redundant information of a physical system, therefore choosing features which are likely to be relevant for model performance and reduce noise. If applyable, this methodology, combined with rather brute methods like correlation matrices, should represent a rather save way of selecting features.
 - Feature creation: Based on the physical context of the problem I guessed that a quantity which is "proportional" to the general intensitiy of the electromagnetic radiation coming from the objects, might me relevant. With the help of this additional feature model performance improved from around 97.7% to practically 100%.

Even though the performance is already close to best here are some ideas for improvements:
 - In the final model, it seems like the 'redshift' feature is not really well picked up by our model, also, most if not all of the falsely predicted objects, have their added feature 'extra_1' really close to 0. Both of these observations could be implemented by additional feature engineering, to potentially further increase model performance.
