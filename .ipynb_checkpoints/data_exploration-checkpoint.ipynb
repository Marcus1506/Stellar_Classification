# Data exploration


```python
import polars as pl
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

## 1) Reading in the data frame:

```python
df=pl.read_csv("data/star_classification.csv", infer_schema_length=300)
# increased detection range for data types accurately, which was necessary in this case
```

## 2) Description from kaggle about the different data types:

The data consists of 100,000 observations of space taken by the SDSS (Sloan Digital Sky Survey). Every observation is described by 17 feature columns and 1 class column which identifies it to be either a star, galaxy or quasar.

obj_ID = Object Identifier, the unique value that identifies the object in the image catalog used by the CAS

alpha = Right Ascension angle (at J2000 epoch)

delta = Declination angle (at J2000 epoch)

u = Ultraviolet filter in the photometric system

g = Green filter in the photometric system

r = Red filter in the photometric system

i = Near Infrared filter in the photometric system

z = Infrared filter in the photometric system

run_ID = Run Number used to identify the specific scan

rereun_ID = Rerun Number to specify how the image was processed

cam_col = Camera column to identify the scanline within the run

field_ID = Field number to identify each field

spec_obj_ID = Unique ID used for optical spectroscopic objects (this means that 2 different observations with the same spec_obj_ID must share the output class)

class = object class (galaxy, star or quasar object)

redshift = redshift value based on the increase in wavelength

plate = plate ID, identifies each plate in SDSS

MJD = Modified Julian Date, used to indicate when a given piece of SDSS data was taken

fiber_ID = fiber ID that identifies the fiber that pointed the light at the focal plane in each observation

- From this description we can identify some of the features which are not necessary for our classification (requires physical intuition).

- This works under the assumption that the are no flaws involved in the process of actually gathering the data.

- Even if we dont fully understand each feature (would require more detailed knowledge about the physical implications at hand) we should later be able to identify further features which arent well suited for usage in the classification (by looking at hit and miss rate for each feature for example).

# 3) Eliminating unnecessary data (1st round)
Features which describe position and orientation relative to the observer are probably of no use for our classification.

The same is also true for different kinds of IDs. (not sure about redshift)

```python
df_cut1=df.drop(["obj_ID", "run_ID", "rerun_ID", "cam_col", "field_ID", "plate", "MJD", "fiber_ID"])
```

```python
df_cut1.describe()
```

```python
df_cut1.filter(df.is_duplicated()) # look for duplicated data
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x=df_cut1.get_column("redshift"), y=df_cut1.get_column("spec_obj_ID"), s=4)
plt.yscale("log")
```

From the context we can infer that "spec_obj_ID" is actually just how far the object is away from the observer.

Alpha and beta are values which just refer to the orientation of the object relative to the observer.

Therefore if we drop the alpha and delta we should have eliminated all unnecessary information.

```python
df_cut2=df_cut1.drop(["alpha", "delta"])
```

```python
df_cut2.filter(df_cut2.is_duplicated()) # check if some information is non unique
```

```python
df_cut2.head()
```

## 4) Changing non numerical datatypes to one-hot encoded datatype

```python
df_cut2.get_column("class").unique()
```

```python
OHE=OneHotEncoder()
```

```python
feature_array=OHE.fit_transform(np.array(df_cut2.get_column("class")).reshape(-1, 1)).toarray()
print(feature_array)
```

```python
target_data=pl.DataFrame(feature_array, schema=["GALAXY", "STAR", "QSO"])
target_data.head()
```

```python
df_cut2_converted=pl.concat([df_cut2, target_data], how="horizontal").drop("class")
df_cut2_converted.head()
```

```python
df_cut2_converted.describe()
```

## 5) Handling data imbalances

```python
plt.hist(x=df_cut2_converted.get_column("u"), bins=20)
plt.show()
plt.hist(x=df_cut2_converted.get_column("g"), bins=20)
plt.show()
plt.hist(x=df_cut2_converted.get_column("r"), bins=20)
plt.show()
plt.hist(x=df_cut2_converted.get_column("i"), bins=20)
plt.show()
plt.hist(x=df_cut2_converted.get_column("z"), bins=20)
plt.show()
plt.hist(x=df_cut2.get_column("class"), bins=3)
plt.show()
plt.hist(x=df_cut2.get_column("spec_obj_ID"), bins=20)
plt.show()
plt.hist(x=df_cut2_converted.get_column("redshift"), bins=20)
plt.show()
```

Remove data imbalances:

```python
df_values_removed=df_cut2_converted.filter((pl.col("u")>=0) | (pl.col("g")>=0) | (pl.col("z")>=0))
df_values_removed.describe()
```

Handling imbalanced label distribution:

```python
df_imbalances_removed=df_values_removed.filter((pl.col("STAR")==1.) | (pl.col("QSO")==1.))
df_GALAXY=df_values_removed.filter(pl.col("GALAXY")==1.).sample(frac=0.35)
df_final=pl.concat([df_imbalances_removed, df_GALAXY], how="vertical")
df_final.describe()
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

## 6) Normalize dataset

```python
X_train, X_test, y_train, y_test = train_test_split(df_final.select(["u","g","r","i","z","redshift","spec_obj_ID"]), df_final.select(["GALAXY","STAR","QSO"]), test_size=0.15, random_state=42)

# because of datatype issues when training model with polars datatype, convert to numpy (this transposes the arrays for some reason):
X_train=np.transpose(X_train) #this gets automatically converted to numpy datatype, so only transpose
X_test=np.transpose(X_test)
y_train=np.transpose(np.array(y_train))
y_test=np.transpose(np.array(y_test))

scaler=preprocessing.StandardScaler().fit(X_train)
X_train=scaler.transform(X_train)

scaler=preprocessing.StandardScaler().fit(X_test)
X_test=scaler.transform(X_test)

print(np.shape(y_train), np.shape(y_test))
print(np.shape(X_test))
print(X_train.shape)
```

```python
print(X_train.mean(axis=0), X_train.std(axis=0))
```

## 7) Build and train model

```python
import tensorflow as tf
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.python.client import device_lib
import sklearn.metrics
```

```python
callback=tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)

model=Sequential()
model.add(Dense(units=7, input_shape=(7,)))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer="Adam" , metrics='categorical_crossentropy')

history=model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=32, epochs=100, shuffle=True, callbacks=[callback])
```

## 8) Analyze model performance
Using confusion matrices and looking at significance of each feature for predictions:

```python
def convert_to_binary(yhat):
    for i in range(len(yhat)):
        max_element=max(yhat[i])
        for j in range(len(yhat[i])):
            if yhat[i,j]<max_element:
                yhat[i,j]=0.
            else:
                yhat[i,j]=1.
    return yhat
```

```python
yhat_test=model.predict(X_test)
yhat_test=convert_to_binary(yhat_test)

cm_GALAXY=sklearn.metrics.confusion_matrix(y_true=y_test[:,0], y_pred=yhat_test[:,0])
dp_GALAXY=sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=cm_GALAXY)
cm_STAR=sklearn.metrics.confusion_matrix(y_true=y_test[:,1], y_pred=yhat_test[:,1])
dp_STAR=sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=cm_STAR)
cm_QSO=sklearn.metrics.confusion_matrix(y_true=y_test[:,2], y_pred=yhat_test[:,2])
dp_QSO=sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix=cm_QSO)

dp_GALAXY.plot()
dp_STAR.plot()
dp_QSO.plot()
```

#### Looking at accuracy and precision:

```python
acc_G=np.trace(cm_GALAXY)/np.sum(cm_GALAXY)
print("Accuracy of identifying Galaxies: ", acc_G)
acc_S=np.trace(cm_STAR)/np.sum(cm_STAR)
print("Accuracy of identifying Stars: ", acc_S)
acc_Q=np.trace(cm_QSO)/np.sum(cm_QSO)
print("Accuracy of identifying Quasars: ", acc_Q)
```

#### Looking at how well the model captures dependencies on some features:

```python
X_hit=[]
X_miss=[]

for i in range(len(yhat_test)):
    if np.array_equal(yhat_test[i], y_train[i]):
        X_hit.append(X_test[i,:])
    else:
        X_miss.append(X_test[i,:])

X_hit=np.array(X_hit)
X_miss=np.array(X_miss)

x_k_hit=X_hit.mean(axis=0)/X_hit.std(axis=0)
x_k_miss=X_miss.mean(axis=0)/X_miss.std(axis=0)

plt.figure(figsize=(6,4), dpi=150)
plt.plot([i for i in range(len(x_k_hit))], x_k_hit, ".--r", label=r"$\overline{x}_{k,hit}/\sigma_{k,hit}$")
plt.plot([i for i in range(len(x_k_hit))], x_k_miss, "x--b", label=r"$\overline{x}_{k,miss}/\sigma_{k,miss}$")
plt.xlabel("k-th feature")
plt.ylabel(r"$\overline{x}_k/\sigma_k$")
plt.legend()
plt.show()
```

This could imply that our "i"-feature does not get picked up well by our model. Feature engineering like performing non linear operations on this feature or combining it with another "badly" recognised feature could enhance the performance of the model.

Since this model already performs pretty well, I will leave it at that.
